# Restricted Boltzmann Machine (RBM)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import BernoulliRBM
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
print("Demonstration Of Restricted Boltzmann Machine (RBM)\n")
```

```python
print("Generate the sample data")
digits = datasets.load_digits(n_class=6)
X_data = digits.data
y_data = digits.target

scalar = MinMaxScaler()
x_data = scalar.fit_transform(X_data)

X_train, X_test, y_train, y_test = train_test_split(X_data,y_data,test_size=0.2,random_state=42)
print(f"Original data shape: {X_train.shape}")
print(f"Number of features:", X_train.shape[1])
```

```python
print("Initialize the training BernoulliRBM\n")
rbm= BernoulliRBM( n_components = 64,
                  learning_rate = 0.01,
                  batch_size = 10,
                  random_state=0,
                  verbose=True)
rbm.fit(X_train)
print("\nRBM Training Complete")
```

```python
print("Extracting features from training data...")
X_features =  rbm.transform(X_train)
print(f"Original data shape: {X_train.shape}")
print(f"Transformed feature shape: {X_features.shape}")
print(f"Original data shape: {X_train.shape[1]} dimension is now represented by {X_features.shape[1]} features")
```

```python
print("Demonstration of reconstruction of reconstruced features")
X_reconstructed =rbm.gibbs(X_train)
print("Original vs Reconstructed Samples:")

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12,5))
fig.suptitle('Original vs Reconstructed Digits (RBM)', fontsize=16)

for i in range(5):
    # Original images on first row
    axes[0, i].imshow(X_train[i].reshape(8,8), cmap=plt.cm.gray_r)
    axes[0, i].set_title(f'Original {i+1}')
    axes[0, i].axis('off')
    
    # Reconstructed images on second row
    axes[1, i].imshow(X_reconstructed[i].reshape(8,8), cmap=plt.cm.gray_r)
    axes[1, i].set_title(f'Reconstructed {i+1}')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```
